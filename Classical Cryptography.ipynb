Check Prime Number 1 by 1

```python
CheckPrime = int(input("Please enter number to check:"))
Count = 0
for i in range(1,CheckPrime):
    Mod = CheckPrime%i
    if Mod == 0 :
       Count += 1
if Count < 2:
    print(CheckPrime, "is prime number")
else:
    print(CheckPrime, "is not prime number")     
```
